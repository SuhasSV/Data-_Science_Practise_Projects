## Which year had the least number of stops?

```python
ri.stop_date.str.slice(0,4).value_counts()
```

```python
combined=ri.stop_date.str.cat(ri.stop_time,sep=' ')
```

```python
ri['Stop_date_time']=pd.to_datetime(combined)
```

```python
ri.head()
```

```python
ri.dtypes
```

```python
ri.Stop_date_time.dt.year.value_counts().sort_values().index[0]
```

## How does drug activity change by time of day?


```python
ri.drugs_related_stop.mean()# .mean()represents percentage of time it is true
```

```python
ri.groupby(ri.Stop_date_time.dt.hour).drugs_related_stop.mean().plot()
```

## Do most stops occur at night?

```python
ri.groupby(ri.Stop_date_time.dt.hour).stop_date.count().plot()
```
